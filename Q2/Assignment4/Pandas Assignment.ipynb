### 11. How many students got A, B, C, F?

```python
Grade = csvData

LenA = Grade[Grade['Grade'].str[0] == 'A']
LenB = Grade[Grade['Grade'].str[0] == 'B']
LenC = Grade[Grade['Grade'].str[0] == 'C']
LenF = Grade[Grade['Grade'].str[0] == 'F']

print(len(LenA) , ', Students got A')
print(len(LenB) , ', Students got B')
print(len(LenC) , ', Students got C')
print(len(LenF) , ', Students got F')
```
